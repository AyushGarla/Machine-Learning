```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

### Pre processing Data
## Change the arrangement of the data to be understandable type for SK Learn






```python
x.shape
```

```python
x = x.reshape(-1,1) # -1 and 1 are arguments to reshape the array from row to column
y1 = y.reshape(-1,1)
```

```python
x.shape
```

### Run the ML Function - Leanier Regression -
# Creat Model
# Fit

```python
reg = LinearRegression() # Y = ax + b # Create Model
reg.fit(x,y1)
```

## Predict : Assign the results of Y = ax + b to Yhat

```python
yhat = reg.predict(x)
```

## We compare the result of regression with original data

```python
plt.scatter(x,y) # Original Data
plt.plot(x,yhat, c='red') # The result of ML-Leaner Regression
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python

metrics.r2_score(y, yhat)
```

```python
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"

raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

# In this model we define 30 % of the data ( from 506) as test data.
# So the training phase of ML is done by 70 % data.
# The function is  "train_test_split"

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.3)
```

**Creat Model** = by Multivariate regression, which  is a technique used to measure the degree to which the  independent variables ( 13 features ) and the dependent variable ( price of house). And they are linearly related to each other. Y ( dependent variable) = a1x1 + a2x2 + a3x3 + ..... anxn

**Fiting Phase**

We use the training proportion of data ( 70%) to make and fit the model.

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

# **Pridection Phase**

# We put the X_test to get the prediction results and then we compare the results with original y_test which we keep to test whether our model is reliable or not

```python
y_pred = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('original prices')
vv = [0,10,20,30,40,50] # reference line to show how these two amounts ( test and predicted) are different  )
plt.plot(vv,vv, color='red')
plt.ylabel('predicted prices')
plt.show()
```
